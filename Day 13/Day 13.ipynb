```python
def lcm(a):
    lcm = a[0]
    for i in a[1:]:
      lcm = lcm*i//gcd(lcm, i)
    return lcm
```

```python
# q2
buses = [bus for bus in inp[1].split(',')]
paddings = {int(bus): -i % int(bus) for i, bus in enumerate(buses) if bus != "x"}
buses = list(paddings)

stack = paddings[buses[0]]
multiple = buses[0]
for i, bus in enumerate(buses[:-1]):
    while stack % buses[i+1] != paddings[buses[i+1]]:
        stack += multiple
    multiple = lcm([multiple, buses[i+1]])
    
print(stack)
```
