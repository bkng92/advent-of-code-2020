```python
from collections import Counter
from itertools import groupby
```

```python
with open('i.txt') as file:
    adapters = file.read().split('\n')
```

```python
adapters = [int(adapter) for adapter in adapters]
adapters.extend([0, max(adapters)+3])
adapters.sort()
```

```python
# make a list differences of each adapters
diffs = []
for x, y in zip(adapters[0::], adapters[1::]): 
    diffs.append(y-x)  

print(Counter(diffs)[1] * Counter(diffs)[3])
```

```python
# distinct arrangements is a function of number of groups of consecutive repeating ones in diffs list, i.e. 
# 4 consecutive ones = 7 combinations,
# 3 consecutive ones = 4 combinations,
# 2 consecutive ones = 2 combinations,

# count number of groups of consecutive repeating ones
count_dups = [sum(1 for _ in group) for k, group in groupby(diffs)if k == 1]
counter = Counter(count_dups)

print(counter)
print(7**counter[4] * 4**counter[3] * 2**counter[2])
```
